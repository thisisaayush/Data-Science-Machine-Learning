```python
import numpy as np
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.show()
```

```python
#Create figure object.
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('linear line')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,.5,.2,.2])
ax1.plot(x,y, color='blue')
ax2.plot(y,x, color='red')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
ax3 = fig.add_axes([.1,.1,.4,.4])
```

```python
fig = plt.figure()
ax3 = fig.add_axes([0,0,1,1])
ax3.plot(x,y,z)
ax3.set_xlabel('x-axis')
ax3.set_ylabel('y-axis')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])

ax1.plot(x,z)
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')

ax2.plot(x,y)
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title('Zoom')
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[0].set_ylabel('z')
axes[0].set_title('xz subplots')

axes[1].plot(z,y, color='green', lw=3, ls='--')
axes[1].set_xlim([10,250])
axes[1].set_ylim([0,100])
axes[1].set_title('zy subplots')
axes[1].set_ylabel('y')
axes[1].set_xlabel('z')

plt.tight_layout()
```
